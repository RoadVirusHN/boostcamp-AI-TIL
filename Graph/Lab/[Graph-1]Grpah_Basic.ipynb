```
# 실습에 필요한 library를 임포트합니다.
import networkx as nx                           # NetworkX : 그래프를 생성, 변경 시각화하기 위한 라이브러리, 비슷한 걸로 Snap.py(속도는 더 빠르고 사용하기 조금 더 불편함)가 있다.
import numpy as np                              # 선형대수를 위한 라이브러리
import matplotlib.pyplot as plt                 # 그림을 그리기 위한 라이브러리
import sys
np.set_printoptions(threshold=sys.maxsize)      
```

```
print("###### Graph Init ######")               
G= nx.Graph()                                   # 방향성이 없는 그래프
DiGraph = nx.DiGraph()                          # 방향성이 있는 그래프
```

```
print("###### Add Node to Graph ######")                    
print("# Add node 1")                                      
G.add_node(1)                                               # 정점 1 추가
print("Num of nodes in G : " + str(G.number_of_nodes()))    # 정점의 수 반환
print("Graph : " + str(G.nodes)+ "\n")                      # 정점의 목록 반환
G.add_node("bruh") # 정점의 명칭은 관계없는듯
print(G.nodes)
```

```
print("# Add vertex 2 ~ 10")                                # 정점 2 ~ 10 추가
for i in range (1, 11):
    G.add_node(i)
print("Num of nodes in G : " + str(G.number_of_nodes()))
print("Graph : " + str(G.nodes) + "\n")
```

```
print("###### Add Edge to Graph ######")                    
G = nx.Graph()
print("#Add edge (1, 2)")                                   
G.add_edge(1, 2)                                            # 정점 1과 2 사이에 간선 추가
print("Graph : " + str(G.edges) + "\n")                     # 간선의 목록 반환
print(str(G.nodes)) # node 또한, 1과 2 생성
```

```
print("#Add edge (1, i) for i = 2 ~ 10")                    # 정점 1과 다른 정점 사이의 간선 추가
for i in range (2, 11):
    G.add_edge(1, i)
print("Graph : " + str(G.edges) + "\n")
```

```
# 그래프를 시각화
# 정점의 위치 결정
pos = nx.spring_layout(G)    # 간선들이 겹치지 않게 이쁘게 펼쳐친 position 그려줌
print(pos)                     
# 정점의 색과 크기를 지정하여 출력
im = nx.draw_networkx_nodes(G, pos, node_color="red", node_size=100)   # 정점의 색깔과 표시될 정점의 크기
# 간선 출력
nx.draw_networkx_edges(G, pos)                                          
# 각 정점의 라벨을 출력
nx.draw_networkx_labels(G, pos, font_size=10, font_color="black")       
plt.show()
```

```
# 그래프를 인접 리스트로 저장
print("인접 리스트: ", nx.to_dict_of_lists(G))

# 그래프를 간선 리스트로 저장
print("간선 리스트: ", nx.to_edgelist(G))

# 그래프를 인접 행렬(일반 행렬)로 저장 #  전체 원소를 저장하므로 정점 수의 제곱에 비례하는 저장 공간을 사용
print("일반 행렬: ", nx.to_numpy_array(G))

# 그래프를 인접 행렬(희소 행렬)로 저장 # 0이 아닌 원소만을 저장하므로 간선의 수에 비례하는 저장 공간을 사용, 메모리 효율적
# 하지만 행렬의 대부분의 원소가 0이 아닐 경우 희소행렬이 더 느리다.
print("희소 행렬: ", nx.to_scipy_sparse_matrix(G))
```
